**`Music`** is a universal language, transcending cultures and time. It is a powerful art form that can evoke a wide range of emotions, from joy and excitement to sadness and reflection. Music can be used to express oneself, to connect with others, and to celebrate life. We are studying this dataset because it could be essential for studying music and developing new music technologies. The dataset is provided as a `.csv` file where it can be viewed in Excel and Notepad. 

This dataset contains 17,996 **rows** across 17 **columns**. Each row represents **1 song**, while columns represent **audio features**. The following are the columns in the dataset and their descriptions:

| Column Name | Description |
| --- | --- |
| **`Artist Name`** | Name of artist |
| **`Track Name`** | Name of song |
| **`Popularity`** | A value between 0 and 100, calculated by an algorithm and is based, in the most part, on the total number of plays the track has had and how recent those plays are |
| **`danceability`** | Describes how suitable a track is for dancing; 0.0 is least danceable and 1.0 is most danceable |
| **`energy`** | A measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity |
| **`key`** | The key the track is in, integers map to pitches using standard [Pitch Class notation](https://en.wikipedia.org/wiki/Pitch_class); -1 if no key was detected |
| **`loudness`** | The quality of a sound that is the primary psychological correlate of physical strength (amplitude), values are averaged across the entire track; in decibels (dB) |
| **`mode`** | Indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived; 1 is Major and 0 is Minor |
| **`speechiness`** | The presence of spoken words in a track; >0.66 is probably made entirely of spoken words, 0.33-0.66 may contain both music and speech, <0.33 most likely represents music |
| **`acousticness`** | A confidence measure from 0.0 to 1.0 of whether the track is acoustic |
| **`instrumentalness`** | Predicts whether a track contains no vocals; >0.5 is intended to represent instrumental tracks |
| **`liveness`** | Detects the presence of an audience in the recording; >0.8 provides strong likelihood that the track is live |
| **`valence`** | A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track; tracks with high valence sound more positive, and vice versa|
| **`tempo`** | The overall estimated tempo of a track in beats per minute (BPM) |
| **`duration_in min/ms`** | Duration in millisecond (ms) |
| **`time_signature`** | A notational convention to specify how many beats are in each bar |
| **`Class`** | corresponds to the genre of the track |

### Question 2: Identify tracks that have high danceability but low energy, and vice versa.

```python
# Set your threshold values for danceability and energy
high_danceability_threshold = 0.8
low_energy_threshold = 0.3

# Identify tracks with high danceability and low energy
high_dance_low_energy_tracks = music_df[
    (music_df['danceability'] > high_danceability_threshold) &
    (music_df['energy'] < low_energy_threshold)
]

# Identify tracks with low danceability and high energy
low_dance_high_energy_tracks = music_df[
    (music_df['danceability'] < high_danceability_threshold) &
    (music_df['energy'] > low_energy_threshold)
]

# Display the results
print("Tracks with High Danceability and Low Energy:")
print(high_dance_low_energy_tracks[['Track Name', 'danceability', 'energy']])

print("\nTracks with Low Danceability and High Energy:")
print(low_dance_high_energy_tracks[['Track Name', 'danceability', 'energy']])
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))

# Plot tracks with high danceability and low energy
sns.scatterplot(
    x='danceability',
    y='energy',
    data=high_dance_low_energy_tracks,
    color='blue',
    label='High Danceability, Low Energy'
)

# Plot tracks with low danceability and high energy
sns.scatterplot(
    x='danceability',
    y='energy',
    data=low_dance_high_energy_tracks,
    color='red',
    label='Low Danceability, High Energy'
)

# Set labels and title
plt.title('Scatter Plot of Danceability vs Energy with Thresholds')
plt.xlabel('Danceability')
plt.ylabel('Energy')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Combine the data for plotting
combined_data = pd.concat([
    high_dance_low_energy_tracks.assign(Group='High Danceability, Low Energy'),
    low_dance_high_energy_tracks.assign(Group='Low Danceability, High Energy')
])

# Bar chart with counts displayed on top of each bar
plt.figure(figsize=(8, 6))

# Plot the bar chart
ax = sns.countplot(x='Group', data=combined_data)

# Display counts on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Set labels and title
plt.title('Comparison of Tracks: High Danceability, Low Energy vs Low Danceability, High Energy')
plt.xlabel('Group')
plt.ylabel('Number of Tracks')

# Display the plot
plt.show()
```

### Question 3: Plot a kernel density estimate (KDE) of both danceability and energy to better understand their distributions.

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Plot KDE for danceability
sns.kdeplot(data=music_df, x='danceability', fill=True, label='Danceability', color='skyblue')

# Plot KDE for energy
sns.kdeplot(data=music_df, x='energy', fill=True, label='Energy', color='salmon')

# Set labels and title
plt.title('Kernel Density Estimates for Danceability and Energy')
plt.xlabel('Value')
plt.ylabel('Density')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
